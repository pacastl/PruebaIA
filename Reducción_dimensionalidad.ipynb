Reducción de dimensionalidad es una técnica que se utiliza para reducir el número de características en un conjunto de datos

Voy a probar PCA (técnica lineal y se basa en la identificación de una combinación lineal de características que representan la mayor cantidad posible de la variabilidad) 

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Seleccionar solo las columnas numéricas
num_cols = ['SL', 'EEG', 'BP', 'HR', 'CIRCULATION', 'ACTIVITY']
data = df[num_cols].values
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_scaled)

plt.scatter(pca_result[:,0], pca_result[:,1])
plt.title('PCA')
plt.show()
```

 t-SNE (no lineal) y se basa en la probabilidad de similitud entre puntos en un espacio de alta dimensión y en un espacio de baja dimensión.

```python
tsne = TSNE(n_components=2, perplexity=30.0)
tsne_result = tsne.fit_transform(data_scaled)

plt.scatter(tsne_result[:,0], tsne_result[:,1])
plt.title('t-SNE')
plt.show()
```
